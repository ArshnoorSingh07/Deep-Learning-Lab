```python
import pandas as pd
df = pd.read_csv("dataset.csv")
df
```

```python
print(df.head())
print(df.columns)
print(df.shape)
```

```python
X = df[["age","experience"]]
y= df["income"]
```

```python
import numpy as np
n_features = X.shape[1]
print(n_features)
w= np.zeros(n_features)
b= 0.0
```

Forward Pass

```python
def predict(X,w,b):
    """
    X: input features
    w: weights
    b: bias
    """ 
    y_pred= X.dot(w)+b
    return y_pred
```

Loss Function

```python
def mean_squared_error(y,y_pred):
    """
    y: actual values
    y_pred: predicted values
    """
    
    loss= ((y_pred-y)**2).mean()
    return loss
```

Compute Gradients

```python
def compute_gradients(X,y,y_pred):
    N= len(y)
    
    #Gradient wrt weights
    dw= (2/N)*X.T.dot(y_pred-y)
    
    #Gradient wrt bias
    db= (2/N)*(y_pred-y).sum()
    return dw,db
```

Update Parameters(Gradient Descent)

```python
def update_parameters(w,b,dw,db,lr):
    w=w-dw*lr
    b=b-db*lr
    return w,b
lr=0.0001
epochs=100000
```

Training Loop

```python
for epoch in range(epochs):
    y_pred= predict(X,w,b)
    loss= mean_squared_error(y,y_pred)
    dw,db= compute_gradients(X,y,y_pred)
    w,b = update_parameters(w,b,dw,db,lr)
    print(y_pred)
    print(w,b)
    print("Loss: ",loss)
```

```python
w_final=w
b_final=b
print(w_final,b_final)
```

Let us store the predictions now

```python
df_pred= df.copy()
df_pred["y_pred"]= y_pred
df_pred.to_csv("result.csv",index=False)
print(df_pred)
```
